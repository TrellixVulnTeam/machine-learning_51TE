## Basic Human Pose Estimation on an Image

Uses pretrained models from Model Zoo to detect people in that image and estimate their poses.

Based off:
https://gluon-cv.mxnet.io/build/examples_pose/demo_alpha_pose.html

```python
from matplotlib import pyplot as plt
from gluoncv import model_zoo, data, utils
from gluoncv.data.transforms.pose import detector_to_alpha_pose, heatmap_to_coord_alpha_pose
```

### Load pretrained model

Get Alpha Pose model trained with input images (256 x 192) on MS COCO dataset. Uses ResNet-101 V1b as the base model from Model Zoo

```python
detector = model_zoo.get_model('yolo3_mobilenet1.0_coco', pretrained=True)
pose_net = model_zoo.get_model('alpha_pose_resnet101_v1b_coco', pretrained=True)

detector.reset_class(["person"], reuse_weights=['person'])
```

### Preprocess image for detector

Download an image and pre-process with preset data transforms

Returns:

NDArray of shape (batch_size, RGB_channels, height, width), can be fed in directly to the model

Images in numpy format to be plotted

```python
im_fname = utils.download('https://github.com/dmlc/web-data/blob/master/gluoncv/pose/soccer.png?raw=true', path='soccer.png')

x, img = data.transforms.presets.yolo.load_test(im_fname, short=512)
print('Shape of preprocessed image:', x.shape)

class_IDs, scores, bounding_boxes = detector(x)
```

### Process tensor from detector to keypoint network

Alpha Pose network expects input of size 256 x 192 and human is centered. Crop the bounding box area for each human, resize and normalize.

Upscale box size to make sure bounding box includes entire person.

```python
pose_input, upscale_bbox = detector_to_alpha_pose(img, class_IDs, scores, bounding_boxes)
```

### Predict with the Alpha Pose network

Alpha Pose network predicts the heatmap for each joint.

```python
predicted_heatmap = pose_net(pose_input)
pred_coords, confidence = heatmap_to_coord_alpha_pose(predicted_heatmap, upscale_bbox)
```

```python
ax = utils.viz.plot_keypoints(img, pred_coords, confidence, class_IDs, bounding_boxes, scores, 
                              box_thresh=0.5, keypoint_thresh=0.2)

plt.show()
```
